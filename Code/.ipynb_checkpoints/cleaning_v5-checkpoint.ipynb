```python
def weather_merge(stadiums, stations):
    #for each lat/long stadium, find the minimum haversine distance and append that weather station
    for index1 in stadiums.index:
        haversines = []
        for index2 in stations.index:
            h = haversine([stadiums.ix[index1,'Latitude'],stadiums.ix[index1,'Longitude']],
                                       [stations.ix[index2,'lat'],stations.ix[index2,'long']])
            haversines.append([h, index2])
        haversines = sorted(haversines, key=lambda x: x[0])
        min_hav = haversines[0]
        min_station = stations.ix[min_hav[1],'station']
        stadiums.ix[index1,'closest_station'] = min_station
    return stadiums
```

```python
def clean_html_goals_shots(element):
    try:
        soup = BeautifulSoup(element)
        text = zip(soup.findAll('subtype'),soup.findAll('team'))
        output = []
        for i in text:
            output.append((i[0].get_text(),i[1].get_text()))
        return output
    except:
        return element
```

```python
final_merge['goal'].map(clean_html_goals_shots)
```

```python
def split_goal_stats(df):
    goal_list = df['goal'].map(clean_html_goals_shots)
    home_full = []
    away_full = []
    for home,lst in zip(df['home_team_api_id'],goal_list):
        home_goals = []
        away_goals = []
        try:
            for i in lst:
                if float(i[1])==home:
                    home_goals.append(i[0])
                else:
                    away_goals.append(i[0])
        except:
            home_goals.append('')
            away_goals.append('')
        home_full.append(home_goals)
        away_full.append(away_goals)
    df['home_goal_types'] = home_full
    df['away_goal_types'] = away_full
    return df
```

```python
def clean_html_cross(element):
    try:
        soup = BeautifulSoup(element)
        text = soup.findAll('team')
        output = {}
        for i in text:
            output[i.get_text()] = 0
        for i in text:
            output[i.get_text()] += 1
        return output
    except:
        return element
```

```python
def split_cross(df):
    cross_dict = df['cross'].map(clean_html_cross)
    home_crosses = []
    away_crosses = []
    for key, id_ in zip(cross_dict,df['home_team_api_id']):
        try:
            for k,v in key.items():
                if float(k) == id_:
                    home_crosses.append(v)
                else:
                    away_crosses.append(v)
        except:
            home_crosses.append('')
            away_crosses.append('')
    df['home_crosses'] = home_crosses
    df['away_crosses'] = away_crosses
    return df
```
