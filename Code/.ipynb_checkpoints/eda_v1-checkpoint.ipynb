```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats

sns.set_style('whitegrid')
plt.style.use('fivethirtyeight')

from ipywidgets import *
from IPython.display import display

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
input_data = pd.read_csv('/home/llevin/Desktop/capstone/cleaned_data/input_data_v1.csv')
```

```python
input_data.head(3).T
```

```python
input_data.drop('away_post.1',axis=1,inplace=True)
input_data.drop('Unnamed: 0',axis=1,inplace=True)
input_data.drop('Latitude',axis=1,inplace=True)
input_data.drop('Longitude',axis=1,inplace=True)
```

```python
input_data.info()
```

```python
input_data.isnull().sum()
```

```python
input_wo_nulls = input_data[~input_data['City'].isnull()]
```

```python
input_wo_nulls.isnull().sum()
```

```python
input_wo_nulls.describe().T
```

```python
def plot_cols(x,y):
    fig = plt.figure(figsize=(10,7))
    ax = fig.gca()
    ax.scatter(input_wo_nulls[x],input_wo_nulls[y],s=50)
    plt.title(x+' Vs. '+y,fontsize=20)
    plt.show()
```

```python
plot_cols('home_shots','away_shots')
```

```python
plot_cols('Attendance','full_time_home_goals')
```

```python
plot_cols('Attendance','full_time_away_goals')
```

```python
plot_cols('home_shots','Attendance')
```

```python
for count in range(0,20):
    x = input_wo_nulls.iloc[:,np.random.random_integers(0,max(enumerate(input_wo_nulls))[0])]
    y = input_wo_nulls.iloc[:,np.random.random_integers(0,max(enumerate(input_wo_nulls))[0])]
    while((((x.dtype!=float)and(x.dtype!=int))or((y.dtype!=float)and(y.dtype!=int)))or(x[50]==y[50])):
        x = input_wo_nulls.iloc[:,np.random.random_integers(0,max(enumerate(input_wo_nulls))[0])]
        y = input_wo_nulls.iloc[:,np.random.random_integers(0,max(enumerate(input_wo_nulls))[0])]
    try:
        fig = plt.figure(figsize=(10,7))
        ax = fig.gca()
        ax.scatter(x,y,s=50)
        plt.title(x.name+' Vs. '+y.name,fontsize=20)
        plt.show()
    except:
        pass
```
