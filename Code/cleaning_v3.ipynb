```python
weather = pd.read_csv('/home/llevin/Desktop/capstone2/cleaned_data/weather_df.csv')
```

```python
weather.head()
```

```python
weather.drop(['Unnamed: 0','long','lat'],axis=1,inplace=True)
```

```python
weather.head()
```

```python
weather['date'] = pd.to_datetime(weather['date'])
weather['date'] = weather['date'].map(lambda x: datetime.datetime(x.year,x.month,x.day,x.hour))
weather['date'].head()
```

```python
print weather.shape
weather.drop_duplicates(inplace=True)
print weather.shape
```

```python
weather['dew (F)'].value_counts()
```

```python
weather = weather.apply(lambda x: x.map(lambda y: 0. if (y=='M' or y==None) else y))
```

```python
weather[['temp (F)','dew (F)','humidity','wind direction','wind speed (mph)','prec (in)','visibility (mi)']] = weather[['temp (F)','dew (F)','humidity','wind direction','wind speed (mph)','prec (in)','visibility (mi)']].apply(lambda x: x.astype(float))
```

```python
weather.info()
```

```python
weather = weather.groupby(['station','date']).mean().reset_index()
```

```python
final_input['date'] = final_input['date'].map(lambda x: x + datetime.timedelta(hours=15))
final_input['date'].head()
```

```python
final_merge = final_input.merge(weather,how='left',left_on=['date','closest_station'],right_on=['date','station'])
```
