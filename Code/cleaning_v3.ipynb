```python
def clean_html(element):
    try:
        soup = BeautifulSoup(element)
        text = zip(soup.findAll('subtype'),soup.findAll('team'))
        output = []
        for i in text:
            output.append((i[0].get_text(),i[1].get_text()))
        return output
    except:
        return element
    
```

```python
final_merge['goal'].map(clean_html)
```

```python
def split_goal_stats(df):
    goal_list = df['goal'].map(clean_html)
    home_full = []
    away_full = []
    for home,lst in zip(df['home_team_api_id'],goal_list):
        home_goals = []
        away_goals = []
        try:
            for i in lst:
                if float(i[1])==home:
                    home_goals.append(i[0])
                else:
                    away_goals.append(i[0])
        except:
            home_goals.append('')
            away_goals.append('')
        home_full.append(home_goals)
        away_full.append(away_goals)
    df['home_goal_types'] = home_full
    df['away_goal_types'] = away_full
    return df
```

```python
split_goal_stats(final_merge)
```
